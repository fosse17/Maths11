## La composition des propositions

Appliqués à des <span style="color:red">propositions</span> élémentaires, les connecteurs construisent des propositions composées.  
Il est possible d'appliquer à nouveau à ces dernières les connecteurs de négation, conjonction, disjonction, implication ou équivalence.  

Ainsi, à partir des propositions :    
p : _"Jacques se promène le samedi"._  
q : _"Jacques se promène le dimanche"._  
r : _"Il pleut"._  

On peut former la proposition composée $\neg r \implies (p \vee q)$ qui se traduira par :  
_"S'il ne pleut pas , Jacques se promène le week-end."_

Pour manipuler une proposition composée, on peut tout simplement parcourir la liste complète
des valeurs de vérité possibles des assertions qui ont servi à la construire. <br>
Ceci permet de remplacer un raisonnement par une simple vérification mécanique, exécutable par ordinateur. <br>
Les tables de vérité  servent de point de départ à ces vérifications.  
Rappel :


**Exemple**. Si $P$ et $Q$ sont deux propositions, alors $P \implies Q$ est équivalente à $(¬P) \vee Q$. En effet,
on peut vérifier cela à l’aide d’une table de vérité :

$$
\begin{array}{|c|c|c|c|c|c|} \hline
P & Q  &  \color{olive}{P\implies Q } &  \neg P  & \color{blue} {\neg P \vee Q } &  \color{olive}{(P \implies Q) }\iff \color{blue}{(\neg P)\vee Q } \\ \hline
0 & 0 & 1 & 1 & 1 &\color{red} 1\\ \hline
0 & 1 & 1 & 1 & 1 & \color{red}1 \\ \hline
1 & 0 & 0 & 0 & 0 & \color{red}1\\ \hline
1 & 1 & 1 & 0 & 1 & \color{red}1\\ \hline
\end{array}
$$ 

Ainsi l’assertion $(P ⇒ Q) \iff ((¬P) ∨ Q)$ est toujours vraie, quelles que soient les valeurs de
vérité rattachées aux assertions P et Q.   
On qualifie un tel énoncé de **tautologie**.

- Une **tautologie** est une proposition composée qui est vraie quelles que soient les valeurs de vérité des propositions simples qui la composent.

$$
\begin{array}{c} \neg(\neg P)\Leftrightarrow P\\ \neg (P\wedge\neg P) \\ (P\wedge Q)\Leftrightarrow(Q\wedge P) \\ (P\vee Q)\Leftrightarrow(Q\vee P) \end{array}
$$ sont des tautologies.



- Une **antilogie** est une proposition composée qui est fausse quelles que soient les valeurs de vérité des propositions simples qui la composent.

### Règles logiques

Il existe en logique un certain nombre de règles qui établissent un calcul des propositions,
semblable au calcul algébrique.  <br>
Après application de l’une de ces règles sur une assertion, on
obtient une assertion équivalente, c’est-à-dire ayant la même valeur de vérité.

Soient $P, Q, R$ trois assertions.
Nous avons les équivalences (vraies) suivantes :

  - $P \iff \neg(\neg P)$
  
  
  - $(P \wedge Q) \iff (Q \wedge P)$
  
  
  - $(P \vee Q) \iff (Q \vee P)$   
  
  
  - $\color{red}{\neg(P \wedge Q)  \iff  (\neg  P)  \vee (\neg Q)}$ loi de De Morgan 
  
  
  - $\color{red}{\neg(P \vee Q)  \iff  (\neg  P)  \wedge (\neg Q)}$ loi de De Morgan 
  
  
  - $\color{blue}{\big(P \wedge (Q \vee R)  \big)   \iff (P \wedge Q) \vee (P \wedge  R)}$ Distributivité
  
  
  - $\color{blue}{\big(P \vee (Q \wedge R)  \big)   \iff (P \vee Q) \wedge (P \vee  R)}$ Distributivité
  
  
  - $P \implies Q$  $\iff$ $\neg(Q) \implies \neg(P)$


**Exemples de démonstrations :**

* Il suffit de comparer les deux assertions  $\neg(P \wedge Q)$  et $(\neg  P)  \vee (\neg Q)$  pour toutes les valeurs possibles de $P$ et $Q$.

Par exemple si $P$ est vrai et $Q$ est vrai alors :

- $P \wedge Q$  est vrai donc $\neg(P \wedge Q)$  est faux ;

- d'autre part ($\neg P$) est faux,  $(\neg Q$) est faux donc $(\neg  P)  \vee (\neg Q)$  est faux.

Ainsi dans ce premier cas les assertions sont toutes les deux fausses.

On dresse ainsi les deux tables de vérités et comme elles sont égales les deux assertions sont équivalentes.

$$
\begin{array}{|c|c|c|c|}\hline
\color{olive}P & \color{olive}Q  & \color{blue}{V \wedge Q} & (\neg  P)  \vee (\neg Q) \\ \hline
\color{olive}{0} & \color{red}{0} & \color{red}{0}& \color{red}{0} \\ \hline
\color{olive}{0}& \color{red}{1}& \color{red}{1}& \color{red}{1} \\ \hline
\color{olive}{1}& \color{red}{0}& \color{red}{1}& \color{red}{1} \\ \hline
\color{olive}{1}& \color{red}{1} & \color{red}{1} & \color{red}{1} \\ \hline
\end{array}
$$

* On fait la même chose mais il y a trois variables : $P$, $Q$, $R$.  
On compare donc les tables de vérité. Elles sont identiques donc les assertions sont équivalentes.

$$
\begin{array}{|c|c|c|c|c|c|c|c|} \hline
\color{olive}P & \color{olive}Q  & \color{olive}R  &  \color{blue}{Q \vee R} &  P \wedge (Q \vee R) & (P \wedge Q) & (P \wedge  R) & (P \wedge Q) \vee (P \wedge  R) \\ \hline
\color{olive}{0} & \color{red}{0} & \color{red}{0}& \color{red}{}  & & & & \\ \hline
\color{olive}{0}& \color{red}{0}& \color{red}{1}& \color{red}{} & & & & \\ \hline
\color{olive}{0}& \color{red}{1}& \color{red}{0}& \color{red}{} & & & & \\ \hline
\color{olive}{0}& \color{red}{1} & \color{red}{1} & \color{red}{} & & & & \\ \hline
\color{olive}{1} & \color{red}{0} & \color{red}{0}& \color{red}{} & & & & \\ \hline
\color{olive}{1}& \color{red}{0}& \color{red}{1}& \color{red}{} & & & & \\ \hline
\color{olive}{1}& \color{red}{1}& \color{red}{0}& \color{red}{} & & & & \\ \hline
\color{olive}{1}& \color{red}{1} & \color{red}{1} & \color{red}{} & & & & \\ \hline
\end{array}
$$



* Par définition, l'implication $P \implies Q$  est l'assertion $(\neg P) \vee Q$.  

Donc l'implication

$\neg(Q) \implies \neg(P)$  est équivalente à
$\neg(\neg(Q)) \vee \neg(P)$

qui équivaut encore à

$Q \vee \neg(P)$  et donc est équivalente à $P \implies Q$.

ce qui donne formellement : 


$$P \implies Q \iff (\neg P) \vee Q$$

donc 


$$
\begin{align*}
\neg(Q) \implies \neg(P) &\iff  \neg(\neg Q) \vee \neg(P) \\
                                       &\iff  Q \vee \neg(P) \\
                                       &\iff  P \implies Q \\
\end{align*}
$$
                              
On aurait aussi pu encore une fois dresser les deux tables de vérité et voir qu'elles sont égales.


## Contraposée  d'une implication

La proposition $P \implies Q$ a pour **contraposée** la proposition si $(\neg Q) \implies (\neg P)$

Une implication et sa contraposée sont vraies en même temps et fausses en même temps. Elles sont **équivalentes**.

Cette équivalence sera utiliser dans les techniques de démonstration

**Exemple:**

La contraposée du théorème de Pythagore est 

Si dans un triangle ABC, on a $AB^2+AC^2\neq BC^2$ alors le triangle ABC n'est pas rectangle en A.

On l’utilise pour montrer qu’un triangle n’est pas rectangle.

### Réciproque d'une implication

La proposition $P \implies Q$ a pour **réciproque** la proposition si $Q \implies P$

Une implication est sa son implication réciproque non pas toujours la même valeur de vérité.

par exemple : 
* $x=1 \implies x^2=1$ est vrai
* $x^2=1 \implies x=1$ est faux
