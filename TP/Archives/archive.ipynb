## Les ensembles 

Set en Python est une structure de données équivalente à des ensembles en mathématiques. Il peut être composé de divers éléments; 
Vous pouvez ajouter et supprimer des éléments d'un ensemble, vous pouvez parcourir les éléments de l'ensemble, vous pouvez effectuer des opérations standard sur les ensembles (union, intersection, différence). 

En outre, vous pouvez vérifier si un élément appartient à un ensemble.

**L'ordre des éléments est sans importance et les éléments sont uniques.** 

```python
A = {1, 2, 3,2,3}
B = set('qwerty')
print(A)
print(B)
```

```python
A = {1, 2, 3}
B = {3, 2, 3, 1}
print(A == B)
```

```python
A = {1, 2, 3}
print(1 in A, 4 not in A)
```

```python
from math import sqrt
import time

def erast(N):
    L=list(range(N+1))
    for i in L:
        L[i]=True
    L[0],L[1]=False,False

    for i in range(int(sqrt(N)+1)):
        if L[i]:
            j=2*i
            while j<=N:
                if L[j]:
                    L[j]=False
                j=j+i
    
    return [i for i,e in enumerate(L) if e]


abs=[];
ord=[];

for p in range(1,6):
    N=10**p;
    abs.append(N)
   
    start = time.time()
    print(erast(N))
    end = time.time()
    elapsed = end - start
    ord.append(elapsed)
    print(f'Temps d\'exécution pour {N}: {elapsed:.4}ms')  
```

```python
def expo(p,n,m):
    liste=[]
    liste.append(p%m)
    for puis in range(1,n+1):
        liste.append((liste[puis-1]*liste[puis-1])%m);
    return liste;

print(expo(5,3,14))
print(expo(17,7,100))
```

```python
def binaire(p):
    bina=[] 
    while p>0:
        p,r=p//2,p%2
        bina.append(r)
    bina.reverse()
    return(bina)

binaire(154)
```

```python
def exponentiation_rapide(p,n,m):
    """calcul de (p^n)%m"""
    binaire_expo=binaire(n)
    binaire_expo.reverse()
    puis=expo(p,len(binaire(n)),m)
    result=1;
    for pu in range(len(binaire_expo)) :
        if binaire_expo[pu]==1:
            result*=(puis[pu])%m
    
    return result%m

exponentiation_rapide(57383,114,19)
```

```python
def dec2bin(n):
    """retourne sous forme de chaine l'écriture en binaire de  l'entier n écrit en base 10"""
    t = ""
    if n==0:
        return "00000000"
    while n>0:
        quotient, reste = n//2,n%2
        #print("%s=%s*%s+%s"%(n,quotient,2,reste))
        t=str(reste)+t
        n=quotient 
    while(len(t)<8):
        t="0"+t
    return t


def convertit_texte_en_binaire(texte):
    chaine=""
    for lettre in texte:
        b=ord(lettre)
        print(b,dec2bin(b))
        chaine+=dec2bin(b)
    print(chaine)
    return chaine

print(dec2bin(84))
binaire =convertit_texte_en_binaire("iut bordeaux")

```

```python
def facteursPremiers(n):
    if n==1: return [1]
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

```

```python
def facteur_premier_factoriser(n):
    L=facteursPremiers(n)
    L_facto=[]
    for nb in set(L):
        L_facto.append([nb,L.count(nb)])
    return L_facto
```

```python
def calcul_PGCD(L):
    S=1;
    for a in L :
        S*=a[0]**a[1]
    return S


def PGCD_facteur(n1,n2):
    L1=facteur_premier_factoriser(n1)
    L2=facteur_premier_factoriser(n2)
    GCD=[]
    for a in L1:
        for b in L2:
            if a[0]==b[0]:
                GCD.append([a[0],min(a[1],b[1])])
    if GCD==[]: GCD.append([1,1])
    return GCD

print(PGCD_facteur(2,9))
print(PGCD_facteur(40,770))
```

## Test de primalité

On donne le petit **théorème de fermat :** 

>si p est un nombre premier alors pour tout $a \in \mathbb{N}, 
\quad 1\leq a <p, \quad a^{p-1}\equiv 1[p]$

Par contre, si on fixe $a$, il se peut que $ a^{p-1}\equiv 1[ p]$ sans que $p$ ne soit premier; un tel nombre est dit *pseudo-premier de Fermat* de base $a$.
