```python
def dec2bin(n):
    s = ""
    if n==0:
        return "00000000"
    while n>0:
        n, r = n//2,n%2
        s=str(r)+s
   # while(len(s)<8):
    #    s="0"+s
    s=s.zfill(8)
    return s


def text2bin(texte):
    chaine=""
    for l in texte:
        b=ord(l)
        chaine+=dec2bin(b)
    return chaine





f = open('test.txt','r')
message = f.read()
#print(message)
#f.close()

## Test des fonctions
file_binaire=text2bin(message)
print(file_binaire)
```
