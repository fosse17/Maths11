### Exercice 1

Soit $(u_n)$ la suite définie par: $u_0=1$ et $u_{n+1}=\dfrac{1}{2}u_n+2$ pour tout $n\geq 0$.

Démontrer par récurrence que pour tout entier $n\geq  0$ on a :  $$u_n=-3\left(\dfrac{1}{2}\right)^n+4$$

Pour tout  entier $n\geq  0$, on note $P_n$ la proposition :  $P_n:u_n=-3\left(\dfrac{1}{2}\right)^n+4$

#### Initialisation 
$P_0$ est vraie car $u_0=1$ et $-3\left(\dfrac{1}{2}\right)^0+4=-3+4=1$

#### hérédité
Supposons que , pour un entier $n \geq 0$ fixé, $P_n$ est vraie et démontrons alors que $P_{n+1}$ est également vraie.

Nous avons 
$\begin{align*}
u_{n+1} & = \dfrac{1}{2}u_n+2 & \\
    &= \dfrac{1}{2}\left( -3\left(\dfrac{1}{2}\right)^n+4 \right)+2 & \text{ par hypothèse de récurence} \\
    & = -3\left(\dfrac{1}{2}\right)^{n+1}+2 +2 & \text{ en développant le } \dfrac12\\
      & = -3\left(\dfrac{1}{2}\right)^{n+1} +4& \\
\end{align*}$ 

La propriété $P_n$ est donc héréditaire.

#### Conclusion

$P_n$ est donc initialisée pour $n=0$ et héréditaire donc vraie pour tout $n \geq 0$

### Exercice 2

Pour tout $n\geq 1$,soit 
$S_n = \sum_{k=1}^{n} (2k-1)^2$

Démontrer par récurrence que pour tout $n \geq 1$, on a : 

$$S_n = \dfrac{n(2n−1)(2n+1)}{3}$$  

Soit $P_n$ la propososition : $P_n : S_n = \dfrac{n(2n−1)(2n+1)}{3}$

Commencons par écrire $P_{n+1}$ pour voir ce que nous voulons obtenir : 
$P_{n+1} : S_{n+1} = \dfrac{(n+1)(2(n+1)−1)(2(n+1)+1)}{3}=\dfrac{(n+1)(2n+1)(2n+3)}{3}$

#### Initialisation
$P_1$ est vraie car $S_1=1$ et $\frac{1(2\times 1−1)(2\times 1 +1)}{3}=1$

#### hérédité

Supposons que , pour un entier $n \geq 1$ fixé, $P_n$ est vraie et démontrons alors que $P_{n+1}$ est également vraie.


$\begin{align*}
S_{n+1} =& \sum_{k=1}^{n+1} (2k-1)^2 &\\
S_{n+1} =& S_n +(2(n+1)-1)^2 &\\
S_{n+1} =& S_n +(2n+1)^2 &\\
S_{n+1} =& \dfrac{n(2n−1)(2n+1)}{3} +(2n+1)^2 & \text{ par hypothèse de récurrence}\\
S_{n+1} =& (2n+1)\dfrac{n(2n−1)+3(2n+1)}{3} \\
S_{n+1} =& (2n+1)\dfrac{2n^2+5n+3}{3} \\
S_{n+1} =& (2n+1)\dfrac{(n+1)(2n+3)}{3} & \text { en developpant } (n+1)(2n+3) \text{ par exemple }\\
S_{n+1} =& \dfrac{(n+1)(2n+1)(2n+3)}{3}
\end{align*}$

La propriété $P_n$ est donc héréditaire.

#### Conclusion

$P_n$ est donc initialisée pour $n=1$ et héréditaire donc vraie pour tout $n \geq 1$


### Exercice 3

Démontrer par récurrence que pour tout entier naturel $n$ : $\quad 6 \mid 3 n^2 + 3n + 6$

Reprendre cette démonstration avec les **congruences**.

Soit $P_n$ la propososition : 
$$ P_n : 6 \mid 3 n^2 + 3n + 6$$ 

On peut traduire cette proposition par 
$$ P_n : \exists k \in \mathbb{N} \:|\: 3 n^2 + 3n + 6= 6k$$

#### Initialisation
$P_0$ est vraie car $6\mid 6$

#### hérédité

Supposons que , pour un entier $n \geq 0$ fixé, $P_n$ est vraie et démontrons alors que $P_{n+1}$ est également vraie.

Nous avons 
$ \exists k \in \mathbb{N} \:|\: 3 n^2 + 3n + 6= 6k$ par hypothése de récurrence.

$$ \begin{align*}
3 (n+1)^2 + 3(n+1) + 6 &= 3n^2+6n+3+3n+3+6 \\
    &=3n^2+3n+6+6(n+1) \\
    &=6k+6(n+1) \\
    &=6K \text{  avec } K=k+n+1 \in \mathbb{N} \\
    \end{align*} $$
    
La propriété $P_n$ est donc héréditaire.

#### Conclusion

$P_n$ est donc initialisée pour $n=0$ et héréditaire donc vraie pour tout $n \geq 0$

### Exercice 4

Résoudre dans $\mathbb{Z}$ : $\quad 1665x + 1035y = 45$

Après avoir cherché PGCD(1665,1035) et simplifié l'équation, on justifiera l'existence de solutions et on sera précis dans la résolution.

En appliquant l'algorithme d'Euclide
$\begin{align*}
1665 &=& 1035 \times 1 +635 \\
1035 &=&1\times 630+405 \\
630&=& 1 \times 405+225 \\
405&=&1\times 225+180 \\
225&=&1 \times 180 +45 \\
180 &=&4\times 45+0
\end{align*}$

Le PGCD est le dernier reste non nul dans cet algorithme donc $PGCD(1665,1035)=45$

Aprés simplification, l'équation de départ est équivalente à 

$$37x+23y=1$$

Cette équation admet des solutions d'après de théorème de Bachet-Bezout (car $PGCD(37,23)=1$)

On cherche donc un couple solution particulière $(u,v)$ à cette équation : 


$\begin{align*}
37 =& 23 \times 1 +14 &     1=&-3\times 23 +5\times (37-23)&=5\times 37-8\times 23\\
23 =&1\times 14+9 &         1=&2\times 14 -3 \times (23-14)&=-3\times 23 +5\times 14\\
14=& 1 \times 9+5 &         1=& 2\times (14-9)-9 &=2\times 14 -3 \times 9\\
9=&1\times 5+4 &          1=& 5-(9-5) &=2\times 5 -9 \\
5=&1 \times 4 +1 &         1=& 5-4 & \\
4 =&4\times 1+0 &  
\end{align*}$

Une solution particulière est donc $(u,v)=(5,-8)$

On recherche maintenant la solution générale de cette équation : 

$\begin{cases}
37\times x+23\times y&=1 \\
37\times 5+23\times(-8)&=1
\end{cases}$

On soustrait ces deux lignes et on obtient : 
$\begin{align*}
37\times(x-5)+23\times (y+8)=&0 \\
37\times(x-5) =&23(-y-8) & (E_1)\\
\end{align*}$

On en déduit que $37 \mid 23\times(-y-8)$ 

or 37 et 23 sont premiers entre eux (PGCD(37,23=1)) donc par application du théorème de **Gauss** $37 \mid (-y-8)$

Il existe donc un entier relatif $k$ tel que $-y-8=37\times k$ soit $y=-37k-8$

En remplaçant dans $(E_1)$ 
$\begin{align*}
37\times(x-5) =&23(-y-8) \\
37\times(x-5) =&23\times 37k \\
x-5 =&23k \\
x=&23k+5 \\
\end{align*}$

L'ensemble des solutions est donc $S=\{ (23k+5,-37k-8), k \in \mathbb{Z}\}$

**Vérification** : $37\times (23k+5)+23\times (-37k-8)=1$

### Exercice 5

3 -  En déduire, suivant les valeurs de $n$, la valeur de $\Delta$.

2 - Avec un tableau de congruence modulo 7

\hline
5n+1 & 1 & 6& 4 & 2 & 0 & 5 & 3 \\
\hline
2n -1 & 6 & 1 & 3 & 5 & 0 & 2 & 4 \\
\hline
\end{array}$$

3 - On en déduit que si  $n \equiv 4[7]$  $a$ et $b$ sont des multiples de $7$ donc dans ce cas $PGCD(a,b)=7$  
Dans les autres cas, $PGCD(a,b)=1$
